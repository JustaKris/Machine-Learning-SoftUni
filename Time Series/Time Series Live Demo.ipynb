# Time Series
## Live Demo


```python
air_quality = pd.read_csv("data/AirQualityUCI.csv", sep = ",", na_values=["-200"])
```

```python
air_quality= air_quality.dropna(how="all")
```

```python
air_quality.head()
```

```python
air_quality.columns
```

```python
air_quality = air_quality[["Date", "Time", "CO_GT", "PT08_S1_CO", "NMHC_GT", "C6H6_GT",
       "PT08_S2_NMHC", "Nox_GT", "PT08_S3_Nox", "NO2_GT", "PT08_S4_NO2",
       "PT08_S5_O3", "T", "RH", "AH", "CO_level"]]
```

```python
air_quality.head()
```

```python
air_quality.describe().T
```

```python
plt.hist(air_quality["PT08_S1_CO"], bins="fd")
plt.show()
```

```python
air_quality["Time"].str.replace(".", ":", regex = False)
```

```python
air_quality["DateTime"] = pd.to_datetime(air_quality["Date"] + " " + air_quality["Time"].str.replace(".", ":", regex = False))
```

```python
air_quality.drop(["Date", "Time"], axis=1)
```

```python
plt.figure(figsize=(18, 5))

plt.scatter(air_quality["DateTime"], air_quality["PT08_S1_CO"], s = 2)

plt.xlabel("Time")
plt.ylabel("PT08.S1")
plt.show()
```

```python
pd.Series(air_quality[air_quality.DateTime > "2005-11-01"].DateTime.dt.date.unique()).sort_values()
```
